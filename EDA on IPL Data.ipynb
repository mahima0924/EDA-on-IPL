```python
#making a bar-ploat for the  top 5players with most man of the match awards
plt.figure(figsize=(8,5))
plt.bar(list(IPL['player_of_match'].value_counts()[0:5].keys()),list(IPL['player_of_match'].value_counts()[0:5]),color="b")
plt.show()
```

```python
#Getting the frequency of the result column
IPL['result'].value_counts()
```

```python
#Finding out the number of toss wins w.r.t each team
IPL['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first
batting_first=IPL[IPL['win_by_runs']!=0]
```

```python
#looking at the head
batting_first.head()
```

```python
#Making a histograme
plt.figure(figsize=(7,7))
plt.hist(batting_first["win_by_runs"],color='g')
plt.title("Distribution of runs")
plt.xlabel("Runs")
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
#making a barplot for top 3 teams with most wins after batting first
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","yellow","orange"])
plt.show()
```

```python
#making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#Extracting the records where a team won batting second
batting_second=IPL[IPL['win_by_wickets']!=0]
```

```python
#looking at the head
batting_second.head()
```

```python
#Making a histograme
plt.figure(figsize=(7,7))
plt.hist(batting_second["win_by_wickets"],bins=20,color='g')
plt.title("Distribution of wickets")
plt.xlabel("wickets")
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting second
batting_second['winner'].value_counts()
```

```python
#making a barplot for top 3 teams with most wins after batting second
plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["purple","blue","red"])
plt.show()
```

```python
#making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
